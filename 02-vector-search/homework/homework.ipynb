#### Question 3

```python
# Chunk the documents.
from gitsource import chunk_documents

chunks = chunk_documents(
    documents,
    size=2000,
    step=1000,
)
```

```python
# Check length
len(chunks)
```

```python
# Now embed every chunk.
contents = [chunk["content"] for chunk in chunks]

X = embedder.encode_batch(contents)
```

```python
# Now compute similarities.
scores = X.dot(v)
```

```python
# Find the highest score.
import numpy as np

idx = np.argmax(scores)

chunks[idx]["filename"]
```

#### Question 4

```python
# MinSearch instead of manual search
from minsearch import VectorSearch
```

```python
# Create the index
vindex = VectorSearch(keyword_fields=["filename"])
```

```python
vindex.fit(X, chunks)
```

```python
query = "What metric do we use to evaluate a search engine?"

query_vector = embedder.encode(query)

results = vindex.search(query_vector)

results[0]["filename"]
```

#### Question 5

```python
# Create a keyword index.
from minsearch import Index
```

```python
index = Index(
    text_fields=["content"]
)
index.fit(chunks)
```

```python
# Vector search:
query = "How do I store vectors in PostgreSQL?"
query_vector = embedder.encode(query)

vector_results = vindex.search(query_vector, num_results=5)
```

```python
# Text search:
text_results = index.search(query, num_results=5)
```

```python
# Extract filenames.
vector_files = {r["filename"] for r in vector_results}
text_files = {r["filename"] for r in text_results}

print("Vector only:")
print(vector_files - text_files)
```

#### Question 6

```python
def rrf(result_lists, k=60, num_results=5):
    scores = {}
    docs = {}

    for results in result_lists:
        for rank, doc in enumerate(results):
            key = (doc["filename"], doc["start"])
            scores[key] = scores.get(key, 0) + 1 / (k + rank)
            docs[key] = doc

    ranked = sorted(scores, key=scores.get, reverse=True)

    return [docs[key] for key in ranked[:num_results]]
```

```python
# Run the query.
query = "How do I give the model access to tools?"
```

```python
# Vector search:
query_vector = embedder.encode(query)

vector_results = vindex.search(query_vector, num_results=5)
vector_results
```

```python
# Keyword search:
text_results = index.search(query, num_results=5)
text_results
```

```python
results = rrf([vector_results, text_results])

results[0]["filename"]
```
